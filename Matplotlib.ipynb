# Matplotlib

```python
import matplotlib.pyplot as plt
```

%matplotlib inline --- allows to see/visualise plots

```python
%matplotlib inline  
```

```python
import numpy as np
x= np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y) #if it does not plot a graph use 'plt.show()'
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
plt.title('TITLE')
```

```python
plt.subplot(1,2,1)#no.of rows,no.of cols,plot number
plt.plot(x,y,'r')

plt.subplot(1,2,2)#no.of rows,no.of cols,plot number
plt.plot(y,x,'b')
```

```python
#Object oriented programming
fig =plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])#left,bottom,width,height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Tittle')
```

```python
fig=plt.figure()

axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

# Matplotlib-2

```python
fig,axes =plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes =plt.subplots(nrows=3,ncols=2)
```

```python
fig,axes =plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')

plt.tight_layout() #takes care of overlapping 
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,4))#figsize takes two arguments width, height
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4),dpi=100)#figsize takes two arguments width, height
#dpi---dots per inch or pixels per inch
axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure(dpi=300)
axes=fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='X Square')
axes.plot(x,x**3,label='X Cube')

axes.legend(loc=0)
```

```python
fig = plt.figure(dpi=300,figsize=(6,6))
axes=fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='X Square')
axes.plot(x,x**3,label='X Cube')

axes.legend(loc=(0.1,0.1)) #location of legend
```

# Matplotlib--Plot Appearence

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])
#can also write RGB HEX code,alpha=transparency,linewidth=lw
ax.plot(x,y,color='red',lw=3,alpha=1,linestyle='-',marker='o',
       markerfacecolor='yellow',markeredgewidth=1,markeredgecolor='green')
```

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',ls='--')

ax.set_xlim([0,1])#shows X @(0,1)
ax.set_ylim([0,2])#shows y @(0,2)
```
